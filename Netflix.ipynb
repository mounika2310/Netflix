### **pre-processing of the data**

To handel Nan values,

director-replacing null value with 'anonymous'.

***Unnesting data***

```python
#unnest director column
dir=df['director'].apply(lambda a: str(a).split(", ")).tolist()
df1=pd.DataFrame(dir,index=df["title"])
df1=pd.DataFrame(df1.stack().reset_index())
df1.drop(['level_1'],axis=1,inplace=True)
df1.rename(columns={0:'director'},inplace=True)
df1['director'].replace({'nan':'Anonymous'},inplace=True)
df1
```

```python
#unnest cast column
cast=df['cast'].apply(lambda a:str(a).split(', ')).tolist()
df2=pd.DataFrame(cast,index=df['title'])
df2=pd.DataFrame(df2.stack().reset_index())
df2.drop(['level_1'],axis=1,inplace=True)
df2.rename(columns={0:'cast'},inplace=True)
df2['cast'].replace({'nan':'Anonymous'},inplace=True)
df2
```

```python
#unnest listed_in column
listed_in=df['listed_in'].apply(lambda a:str(a).split(', ')).tolist()
df3=pd.DataFrame(listed_in,index=df['title'])
df3=pd.DataFrame(df3.stack().reset_index())

df3.rename(columns={0:'listed_in'},inplace=True)
df3.drop(['level_1'],axis=1,inplace=True)
df3
```

```python
#unnest country column
country=df['country'].apply(lambda a:str(a).split(', ')).tolist()
df4=pd.DataFrame(country,index=df['title'])
df4=pd.DataFrame(df4.stack().reset_index())
df4.rename(columns={0:'country'},inplace=True)
df4.drop(['level_1'],axis=1,inplace=True)
df4['country'].replace({'nan':np.nan},inplace=True)
df4.dropna(inplace=True)

#drop duplicates
df4.drop_duplicates(inplace=True)
df4
```

```python
#merge df1,df2,df3,df4
df5=df2.merge(df1,on=["title"],how="inner")
df6=df5.merge(df3,on=['title'],how='inner')
df_unnested=df6.merge(df4,on=['title'],how='inner')
df_unnested
```

```python
df.info()
```

```python
df=df_unnested.merge(df[['show_id','type','title','date_added','release_year','rating','duration','description']],how='inner',on='title')
df
```

# ***Based on observation,dataset cane be divides into two sections based on content type-Movies/TV Shows***

```python
movies=df.loc[df["type"]=="Movie"]
tv=df.loc[df["type"]=="TV Show"]
movies_per=round(movies.shape[0]/df['type'].shape[0]*100,2)
tvshows_per=round(tv.shape[0]/df['type'].shape[0]*100,2)

```

```python
plt.pie([movies_per,tvshows_per],labels=["Movies","TV Shows"],shadow=True)
plt.axis('equal')
plt.title("percentage of movies and tvshows")
plt.legend(['movies','TV Shows'])
```

The percentage of movies is more compared to tv shows in the dataset coneys the idea that Netflix tends to add more movies to the platform than tv shows,supposingly based on consumers preferencies.There are more than 5000+movies and nearly 2000+Tv shows in the set.

***Heatmap***

```python
df['year']=pd.DatetimeIndex(df['date_added']).year
df['month']=pd.DatetimeIndex(df['date_added']).month
```

Observations:
